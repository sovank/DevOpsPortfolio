# Q2. Deploying a Service on Docker Overlay Network

image.png

```python
docker swarm init

docker network create -d overlay my_overlay_network

docker service create --name overlay_service --network my_overlay_network replicas 1 nginx
```

image.png

```python
docker service scale overlay_service =3

docker service ls
```
